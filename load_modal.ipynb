```python
import tensorflow as tf
from tensorflow.keras.models import load_model
import joblib
import pandas as pd
import numpy as np
```

```python
#load modal,scaler,encoder
%cd /content/drive/MyDrive/Data_Science/deep_learning_projects/ann_curn_modelling
model=load_model('ann_model.h5')
scaler=joblib.load('scaler.pkl')
gender_encoder_lb=joblib.load('gender_encoder.pkl')
geo_encoder_ohe=joblib.load('geo_encoder.pkl')
```

```python
#example input data
input_data ={
    'CreditScore':600,
    'Geography':'France',
    'Gender':'Male',
    'Age':40,
    'Tenure':3,
    'Balance':60000,
    'NumOfProducts':2,
    'HasCrCard':1,
    'IsActiveMember':1,
    'EstimatedSalary':50000
}
```

```python
input_data=pd.DataFrame(input_data,index=[0])
input_data
```

```python
# One-hot encoder 'Geography'
geo_encoded=geo_encoder_ohe.transform([input_data["Geography"]])
geo_encoded_df=pd.DataFrame(geo_encoded,columns=geo_encoder_ohe.get_feature_names_out())
```

```python
#merge geo_encoded
input_data=pd.concat([input_data,geo_encoded_df],axis=1)
input_data.drop(columns=['Geography'],inplace=True)
input_data
```

```python
input_data['Gender']=gender_encoder_lb.transform([input_data['Gender']])
input_data
```

```python
final_data=scaler.transform(input_data)
final_data
```

```python
prediction=model.predict(final_data)
```
